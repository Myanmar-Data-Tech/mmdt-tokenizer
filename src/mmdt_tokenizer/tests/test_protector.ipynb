```python
from mmdt_tokenizer.core import MyanmarSyllableTokenizer
from mmdt_tokenizer.utils.config import DATA_DIR, OUTPUT_DIR

import pandas as pd
from pathlib import Path
```

```python
def test_numeric_expressions(tokenizer):
    text = "တစ်ထောင် နှင့် ၂၀ ကျပ်၊ထည့်သွင်းပြီး 09-123456789 (သို့) ၀၉-၉၀၂၇၈၃၄၅၆ကို ဖုန်းဆက်မည်။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)


def test_abbreviated_organization_names(tokenizer):
    text = "U.N. နှင့် အ.ထ.က (အထက) တို့မှ အစီရင်ခံစာကို ထုတ်ပြန်သည်။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)


def test_date_in_words(tokenizer):
    text = "၂၀၂၅ ခုနှစ် မေလ(၅)ရက် တွင် စတင်မည်။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)


def test_time_expressions(tokenizer):
    text = "မနက် ၁၀:၃၀ နာရီတွင် တွေ့ကြမည်။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)

def test_email_addresses(tokenizer):
    text = "ဆက်သွယ်ရန် admin@example.com သို့ အီးမေးလ် ပို့ပါ။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)


def test_named_entities_with_at_and_parli(tokenizer):
    text = "@ayechan ၏ မှတ်ချက်အရ သမ္မတ သည် ကြေညာခဲ့သည်။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)


def test_urls(tokenizer):
    text = "အသေးစိတ်ကို https://example.com/mypath တွင် ကြည့်ပါ။"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)

def test_parli_name(tokenizer):
    text = "သရေခေတ္တရာ သံယောဇဥ်၊ သမ္မတ ဦး၀င်းမြင့်"
    tokens = tokenizer.word_tokenize(text)
    print(tokens)
```

```python
from mmdt_tokenizer import MyanmarTokenizer
tokenizer = MyanmarTokenizer()


test_numeric_expressions(tokenizer)
test_abbreviated_organization_names(tokenizer)
test_date_in_words(tokenizer)
test_time_expressions(tokenizer)
test_email_addresses(tokenizer)
test_named_entities_with_at_and_parli(tokenizer)
test_urls(tokenizer)
test_parli_name(tokenizer)

```
