```python
from mmdt_tokenizer import MyanmarTokenizer
from mmdt_tokenizer.utils.config import DATA_DIR, OUTPUT_DIR
from pathlib import Path


tokenizer = MyanmarTokenizer()
```

```python
def test_merge_month_days(tokenizer):
    # Case 1: Day/MONTH + [PUNCT] + CL + CLEP -- Combine to one phrase, (Tag: D/M, DAYCL/MCL)
    # Case 2: Day/MONTH -- keep it as if 
 
    text = [["ရှေ့အပတ် ကြာသပတေးနေ့တွင် ကြာပန်း ၅ ခိုင်စီ ယူလာကြပါ။", "ကဆုန် လပြည့်နေ့တွင် ရေသပ္ပာယ်မည်။"], 
            ["ယခုနှစ် ဧပြီလတွင်ကျောင်းပိတ်ရက် ကိုးရက် ရပါသည်။ဆရာမတို့ သစ်ပင်များ စိုက်ပျိုးကြမည်။ ပန်းပင် ၅ပင် ယူလာပါ။"],
            ["တနင်္ဂနွေ၊ တနင်္လာ ချိုးကလေး ကူသံသာ၊ အင်္ဂါ ဗုဒ္ဓဟူး ချိုးကလေး တကူးကူး၊"],
            ["ကြာသပတေး၊သောကြာ ချိုးကလေး မှန်မှန်လာ၊ စနေမှာ မလာအား၊ ချိုးကလေး ပျံလို့သွား"]]
            
    tokens = tokenizer.word_tokenize(text[2])
    print(tokens)

test_merge_month_days(tokenizer)
```

```python
def test_merge_numbers(tokenizer):
    # Case 1: NUMBER or word Number + [PUNCT] + CL -- Combine to one phrase (Tag: Num + NUMCL)
    # Case 2: Word Number + [PUNCT] + CL -- Combine to one phrase (Tag: Num + NUMCL)
    # Case 3: PERCENT WORD --> INUMCL
    # Case 4: WORD NUMBER in other context -- Context (Tag: RAW)
 
    text = [["လူဦးရေ ၁၉၆,၀၀၀ အနက် ရာခိုင်နှုန်း၃၀မှာ အသက် ၁၅ နှစ်အောက် ဖြစ်ကြသည်။"],
            ["တစ်ရာ သုံးဆယ့် ငါးဦးမှာ မိန်းမများ ဖြစ်ပြီး သုံးဆယ် ရာခိုင်နှုန်းသည် ကလေးများ ဖြစ်သည်"],
            ["တစုံတရာကို နှစ်ခြိုက်မိသည့် စားသုံးသူများကို လေးစားပါသည်။ "],
            ["၂၀၂၅ခုနှစ်အတွင်းစပါးသီးနှံထွက်ရှိမှုကို ကြေငြာသွားပါမည်။"]]
            
    tokens = tokenizer.word_tokenize(text)
    print(tokens)

test_merge_numbers(tokenizer)
```

```python
def test_merge_predicate(tokenizer):
    # Case 1: unseen + VEP + SFP -- split to raw , pred if more than 5, else keep it as one pred
    # Case 2: unseen + Conj + [unseen]+ [VEP] + SFP
    # Case 3: Negatiave Phrase -- Neg_Prefix + [VEP]+[VEP]+... + NEG_SFP
    # Case 4: Question Phrase -- Que_Presfix + [VEP]+[VEP]+... + QUE_SFP
    # Case 5: SFP in the middle of the sentence. 
 
    text = [["အကြိမ်ရေစိပ်လာပါတယ်","စုစည်း တက်ရောက် ခဲ့ကြပါသည်။","အရှင်ဖမ်းမိခဲ့တဲ့ တိုက်ပွဲဖြစ်သည်","ရောက်ရှိခဲ့ကြပါတယ်။"],
            ["ဝန်ထမ်းတွေလည်းအဲဒီအချိန်ကတည်းကစက်ရုံကိုစွန့်ခွာသွားတာဖြစ်ပါတယ်။", "ကြိုးစားသော်လည်း စာမေးပွဲကျပါသည်။သို့သော် စိတ်မပျက်ပါ။"],
            ["သူမကို မည်သူမျှမသိပါဘူး။", "အစီအစဥ်ကိုသူမစိတ်ဆိုး မနေပါဘူး", "ဖမ်းဆီးသွားခြင်း မရှိပါဘူး။",
              "စိတ်မ၀င်စားပါ",  "အိမ်မက် မမက်ပါနဲ့"],
            ["ပူးပေါင်းမှု ဘာသဘောဆောင်လဲ", "သူတို့ ဘာကြောင့်အသက်ဆုံးရှုံးခဲ့ရသလဲ"],
            ["ဒေါ်အောင်ဆန်းစုကြည်၏အသက် ၇၀ ပြည့်မွေးနေ့ပွဲကို ခမ်းခမ်းနားနားကျင်းပသည်။", "ပျော်ပျော်ရွှင်ရွှင် စားသုံးရန်မှာကြားသည်။", 
             "ပျော်ရွှင် ခဲ့မှာပါ။နားလည် ကြမှာပါ။"]]
            
    tokens = tokenizer.word_tokenize(text[1][1])
    print(tokens)

test_merge_predicate(tokenizer)
```

```python
import pandas as pd
def test_tagging_csv(tokenizer):
    """Tests the CSV loading/saving feature exposed by the main tokenizer."""
    csv_input_path = DATA_DIR /"test_data.csv" 
    csv_text_path = OUTPUT_DIR / "result_word_text.csv"
    csv_output_path = OUTPUT_DIR / "result_word_tag.csv"

    df = pd.read_csv(csv_input_path)
    
    tokenizer.word_tokenize(df,column = 'original_sentence', 
                                     save_csv=str(csv_text_path), save_tag=str(csv_output_path), 
                                     conll_style=False)

    assert Path(csv_output_path).exists()

test_tagging_csv(tokenizer)
```
